```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
color_img = cv2.imread('/Users/mmoran/Desktop/SetCV/public/set_sample.jpg',1)
gray_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
```

```python
gray_img.shape
```

```python
plt.imshow(gray_img, cmap="gray",interpolation='bilinear')
```

```python
blurred_gray = cv2.GaussianBlur(gray_img,(1,1),0)
ret, thresh = cv2.threshold(blurred_gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh,cmap='gray',interpolation='bilinear')
```

```python
# ret, thresh = cv2.threshold(new_gray, 127, 255, cv2.THRESH_BINARY)
ad_thresh = cv2.adaptiveThreshold(blurred_gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,15,2)
im2, contours, hierarchy = cv2.findContours(ad_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# color = cv2.cvtColor(thresh,cv2.COLOR_GRAY2BGR)
pot_cards = []
components = zip(contours,hierarchy[0])
for component in zip(contours,hierarchy[0]):
    cnt = component[0]
    hrc = component[1]
    x,y,w,h = cv2.boundingRect(cnt)
#     if w > 100 and h > 100 and h < 200:
    if (w / h > 0.6) & (w / h < 0.7) & (w > 0.1 * color_img.shape[0]):
        pot_cards.append(component)
        print("width: {} height: {} ratio: {}".format(w,h,w/h))
        cv2.rectangle(color_img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
x,y,w,h = cv2.boundingRect(pot_cards[9][0])
sample_card = ad_thresh[y:y+h,x:x+w]
color_sample_card = cv2.cvtColor(sample_card,cv2.COLOR_GRAY2BGR)
# plt.imshow(sample_card,cmap='gray',interpolation='bilinear')
im2, contours, hierarchy = cv2.findContours(sample_card, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
shapes = []
for component in zip(contours,hierarchy[0]):
    cnt = component[0]
    hrc = component[1]
    x,y,w,h = cv2.boundingRect(cnt)
#     if w > 100 and h > 100 and h < 200:
    if (h > 0.2 * sample_card.shape[0]) & (w/h > 1.9) & (w/h < 2.1):
        shapes.append(component)
        print("width: {} height: {} ratio: {}".format(w,h,w/h))
        cv2.drawContours(color_sample_card, [cnt], 0, (255,0,0), 3)
#         cv2.rectangle(color_sample_card,(x,y),(x+w,y+h),(0,255,0),2)
print("SHAPE COUNT: {}".format(len(shapes)))
plt.imshow(color_sample_card,cmap='rainbow',interpolation='bilinear')
```
